```python
import pandas as pd
import numpy as np
from catboost import  Pool
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from catboost import CatBoostClassifier
from sklearn.metrics import f1_score
```

```python
dfff = pd.read_csv('train_data.csv')
```

```python
y_test_object = dfff['object']
y_test_relevant = dfff['is_relevant']
y_test_positive = dfff['is_positive']
```

```python
df_embed_object_test = pd.read_csv('testing_loader_is_obj.csv')
df_embed_object_test = df_embed_object_test.drop(['Unnamed: 0'], axis=1)

df_embed_is_relevant_test = pd.read_csv('testing_loader_is_relevant.csv')
df_embed_is_relevant_test = df_embed_is_relevant_test.drop(['Unnamed: 0'], axis=1)

df_embed_is_positive_test = pd.read_csv('testing_loader_is_positive.csv')
df_embed_is_positive_test = df_embed_is_positive_test.drop(['Unnamed: 0'], axis=1)
```

```python
df_embed_object = pd.read_csv('training_loader_is_obj.csv')
df_embed_object = df_embed_object.drop(['Unnamed: 0'], axis=1)

df_embed_is_relevant = pd.read_csv('training_loader_is_obj.csv')
df_embed_is_relevant = df_embed_is_relevant.drop(['Unnamed: 0'], axis=1)

df_embed_is_positive = pd.read_csv('training_loader_is_obj.csv')
df_embed_is_positive = df_embed_is_positive.drop(['Unnamed: 0'], axis=1)
```

```python
data = pd.read_csv("itog_data.csv")
objectt = data['object']

data = pd.read_csv("itog_data.csv")
relevant = data['is_relevant']

data = pd.read_csv("itog_data.csv")
positive = data['is_positive']
```

```python
from sklearn.utils import compute_class_weight
```

# object

```python
class_weights_ob = compute_class_weight(class_weight = 'balanced', classes = np.unique(objectt),y = objectt)
```

```python
class_weights_ob
```

```python
model_objectt = CatBoostClassifier(iterations=3000,
                                    learning_rate=0.01,
                                    depth=2,
                                    eval_metric='MultiClass',
                                    random_seed=0,
                                    class_weights= class_weights_ob
                                    # early_stopping_rounds=50
                                    )
```

```python
train_pool_ob = Pool(df_embed_object, objectt)
val_pool_ob = Pool(df_embed_object_test, y_test_object)


model_objectt.fit(train_pool_ob, eval_set= val_pool_ob, plot=True)
```

```python
val_preds_b = model_objectt.predict(df_embed_object_test)
```

```python
from sklearn.metrics import f1_score
f1_val_ob = f1_score(y_test_object, np.array(val_preds_b).ravel(), average='micro')  
#    'macro'  'weighted'     
print("F1 Score:", f1_val_ob)
```
